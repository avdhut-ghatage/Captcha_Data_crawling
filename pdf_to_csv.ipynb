```python
db =[]
for i in range(1,11):
    doc = PyPDF2.PdfReader(f'temp ({i}).pdf')
    page=doc.pages[0]
    text = page.extract_text().split('\n')
    data = {}
    
    data['S.R.O']=(' ').join(text[8].split(' ')[2:-4])
    data['Date']=text[8].split(' ')[-1]
    data['Document No & Year']=text[9].split(':')[1].strip()
    data['Date of Registration']=text[25].strip()
    for i in range(len(text)):
        
        if '-Market Value' in text[i].split('/'):
            data['Market Value']=text[i+1].split(' ')[1].split('/')[0]
        
        try:
            if text[i].split('/')[1]=='-Market Value':
                data['Consideration Value']=text[i].split('/')[0].split(' ')[-1]
        except:
            pass
        if 'Extent' in text[i].split('/')[0].split('-')[-1]:
            data['Survey No-Extent']=text[i].split(':')[-1].strip()
            
        if '-PR Number' in text[i].split('/'):
            data['PR Number']=text[i+1].strip()
            
        if 'Property Type' in text[i].split('/'):
            data['Property Type']=text[i].split(':')[-1].strip()
            
        if 'Village & Street' in text[i].split('/'):
            data['Village & Street']=text[i].split(':')[-1].strip()
            
        if 'Building Name' in text[i].split('/'):
            data['Building Name']=text[i].split(':')[-1].strip()      
            
        if 'New Door No.' in text[i].split('/'):
            data['New Door No']=text[i].split(':')[-1].strip() 
            
        if 'Old Door No.' in text[i].split('/'):
            if 'Floor' in text[i].split(':')[1].split(' '):
                data['Old Door No']=text[i].split(':')[1].split(' ')[1].strip() 
            else:
                data['Old Door No']=text[i].split(':')[-1].strip()
        if 'Floor No.' in text[i].split('/'):
            data['Floor No']=text[i].split(':')[-1].strip()   
            
        if 'Plot No.' in text[i].split('/'):
            data['Plot No']=text[i].split(':')[-1].strip()
            
        if 'Flat No.' in text[i].split('/'):
            data['Flat No']=text[i].split(':')[-1].strip()
         
    db.append(data)
```

```python
file=pd.DataFrame(db)
file['Market Value']=file['Market Value'].str.replace('Number','')
file
```

```python
file.to_csv('pdf_to_csv.csv',index=False)
```
